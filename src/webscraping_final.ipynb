```python
from webscraping import get_rss, xml_to_df

import pandas as pd 
```

This project is to scrape all newspapers from top college newpaper websites. I will be looking mainly at RSS feeds by looking into links that contains: `/rss` or `/feed` or `/api`.

The list of colleges are as follows:

|**University Name** | **Main School Newspaper Website** |
|---------------------------|---------------------------------------|
| Princeton University      | https://www.dailyprincetonian.com/    |
| MIT                       | https://thetech.com/                  |
| Harvard University        | https://www.thecrimson.com/           |
| Stanford University       | https://stanforddaily.com/            |
| Yale University           | https://yaledailynews.com/            |
| University of Chicago     | https://chicagomaroon.com/            |
| Duke University           | https://www.dukechronicle.com/        |
| Johns Hopkins University  | https://www.jhunewsletter.com/        |
| Northwestern University   | https://dailynorthwestern.com/        |


From the list, I will be scraping newpapers from: **Princeton University** and **Stanford University** and returning it as a csv format.

```python
# all urls
urls = {
    'Princeton' : 'https://www.dailyprincetonian.com/',
    'MIT': 'https://thetech.com/',
    'Harvard': 'https://www.thecrimson.com/',
    'Stanford': 'https://stanforddaily.com/',
    'Yale': 'https://yaledailynews.com/',
    'U Chicago': 'https://chicagomaroon.com/',
    'Duke': 'https://www.dukechronicle.com/',
    'JHU': 'https://www.jhunewsletter.com/',
    'Northwestern': 'https://dailynorthwestern.com/'
}
```

```python
rss = {}
for url in urls:
    v = get_rss(urls[url]) if get_rss(urls[url]) else None
    rss[url] = v
```

No RSS feed from Harvard website:

```python
xml_to_df(rss['Harvard'])
```

Yale's RSS feed is in JSON format.

```python
xml_to_df(rss['Yale'])
```

```python
princeton_rss_df = xml_to_df(rss['Princeton'])
stanford_rss_df = xml_to_df(rss['Stanford'])
```
